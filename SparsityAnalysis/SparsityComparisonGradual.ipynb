```python
activation = "CompTanh_gradual"
try:
    with open("SparsityTest_power{}.json".format(activation), 'r') as nsa:
        data = json.load(nsa)
        print(len(data))
except:
    with open("SparsityTest_power{}0.json".format(activation), 'r') as nsa:
        data = json.load(nsa)
        print(len(data))
    with open("SparsityTest_power{}1.json".format(activation), 'r') as nsa:
        data2 = json.load(nsa)
        print(len(data2))
        data.update(data2)

'''
try:
    with open("SparsityTestAll.json", 'r') as nsa:
        data = json.load(nsa)
        print(len(data))
        
except:
    with open("SparsityTest_prune.json", 'r') as nsa:
        data = json.load(nsa)
        #print(data.keys())
        print(len(data))
    with open("SparsityTest_prune.json", 'r') as nsa:
        data2 = json.load(nsa)
        #print(data2.keys())
        print(len(data2))
        data.update(data2)
    with open("SparsityTest_graft.json", 'r') as nsa:
        data2 = json.load(nsa)
        #print(data2.keys())
        print(len(data2))
        data.update(data2)
    with open("SparsityTest_graft.json", 'r') as nsa:
        data = json.load(nsa)
        #print(data.keys())
        print(len(data))
        data.update(data2)
    with open("SparsityTest_random.json", 'r') as nsa:
        data2 = json.load(nsa)
        #print(data2.keys())
        print(len(data2))
        data.update(data2)
    with open("SparsityTest_random.json", 'r') as nsa:
        data2 = json.load(nsa)
        #print(data2.keys())
        print(len(data2))
        data.update(data2)

    with open("SparsityTestAll.json", 'w') as f:
        json.dump(data, f)
'''
        
acc_lim = [70, 85]
iqr_lim = [2, 15]

def str_to_list(in_string):
    temp = in_string[1:-1].split(',')
    return [int(s) for s in temp]

score_num = 0
data = {t:d for t,d in data.items() if type(d) is dict}
for t,d in data.items():
    
    d["shape"] = str_to_list(d["shape"])
    d["hidden"] = len(d["shape"]) - 2
    method = "_".join(t.split('_')[1:3])
    if method[-4] != '1':
        method = method[:-3] + '0' + method[-3:]
    d["method"] = method
    #d["method"] = t.split(']')[1]
    #d["method"] = d["method"][:re.search(r"\d",d["method"]).start()]
    
    d["idealacc"] *= 100
    acc = d["accs"]
    scores = (list(acc.values()))
    scores = np.vstack([np.array(s) for s in scores])[score_num]
    d["medians"] = np.median(scores) * 100
    d["iqr"] = (np.quantile(scores, 0.75) - np.quantile(scores, 0.25)) * 100
    #for i in range(d["hidden"]):
    #    d["shrink"] = d.get("shrink", 1) * (d["shape"][i] / d["shape"][i + 1]) ** (d["hidden"] + 1 - i)
    #d['loc'], d['spread'] = stats.gumbel_r.fit(scores)
    d["u75"] = scores.max() - np.quantile(scores, 0.25)
    d["synapses"] = sum([n * d["shape"][i + 1] for i,n in enumerate(d["shape"][:-1])])
    #print(d)
    #input()
```

```python
df = pd.DataFrame.from_dict(data, orient = "index")
df.sort_values(["method"], axis = 0, inplace = True)
#print(df.head())
print(df["hidden"].value_counts())
if "computer" in df.columns:
    df.drop(["computer"], axis = 1, inplace = True)
    df.dropna(how = "all", inplace = True)
#df.head()
```

```python
# Plot median accuracy vs num hidden layers
fig = plt.figure()
#fig.set_figheight(1.5)
#fig.set_figwidth(1.75)
df_graft = pd.concat((df[df["method"].str.startswith("graft")],df[df["method"].str.startswith("random")]))
plt0 = sns.boxenplot(x = "hidden", y = "idealacc", data = df_graft, hue="method", palette =sns.color_palette("Paired")) #, palette = sns.cubehelix_palette(8, dark=0.35))
plt.ylabel("Ideal\nAccuracies (%)", labelpad = 0)
plt.xlabel("No. Hidden Layers", labelpad=0)
plt.ylim([60,95])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt0.tick_params(axis='both', which='major', pad=0)
fig = plt0.get_figure()
figname = plots + "VAD_CompTanh_median.pdf"
#fig.savefig(figname, dpi=300, bbox_inches='tight')
```
