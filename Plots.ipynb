```python
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np
import math
import seaborn as sns
import matplotlib.pyplot as plt
import copy

sns.set_style(style = 'white',rc={'figure.figsize':(3.25,1), 'font.family':"Times New Roman", 'font.size': 12,
                                  'patch.edgecolor': 'grey', 'dpi': 300})
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x_max = 7.5
x_min = -x_max
y_max = 1.2
y_min = -y_max

mean0 = -8.0 
std0 = 5.0
mean1 = 8.0 
std1 = 5.0

x = np.linspace(x_min, x_max, 100)
y_tanh = np.tanh(x)
y_comp = 0.5 * np.tanh(x * 0.5)
y_sig = sigmoid(x)


y = 0.5 * (scipy.stats.norm.pdf(x,mean0,std0) + scipy.stats.norm.pdf(x,mean1,std1))

plt.figure(figsize=(1.8,1.8))
plt.plot(x,y_tanh, '-', color='k')
#plt.plot(x,y_comp, color='k',)
#plt.plot(x,y_sig, color='k')
#plt.text(-7.3,-0.9, "$tanh(x)$", fontsize = 10)
#plt.text(-7.3,-0.4, r"$\frac{1}{2}tanh(x/2)$", fontsize = 10)
#plt.text(-7.3,0.1, "$sigmoid(x)$", fontsize = 10)

plt.grid()

plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)

#plt.title('OtherActivation Functions',fontsize=14)

plt.xlabel('Input $x$ (no units)', fontsize = 12)
plt.ylabel('Output $y$ (no units)', fontsize = 12)

plt.axhline(0, ls=':',color='k')
plt.axvline(0, ls=':', color='k')

plt.savefig("TanhFunction.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x_max = 5
x_min = -x_max
y_max = 1.2
y_min = -y_max

mean0 = -8.0 
std0 = 5.0
mean1 = 8.0 
std1 = 5.0

x = np.linspace(x_min, x_max, 100)
y_tanh = np.tanh(x)
y_comp = 0.5 * np.tanh(x * 0.5)
y_sig = sigmoid(x)


y = 0.5 * (scipy.stats.norm.pdf(x,mean0,std0) + scipy.stats.norm.pdf(x,mean1,std1))

plt.figure(figsize=(1.8,1.8))
plt.plot(x,np.append([0]*50,x[50:]), color='k')

plt.grid()

plt.xlim(x_min,x_max)
plt.ylim(-0.5,5)

#plt.title('Relu Function',fontsize=14)

plt.xlabel('Input $x$ (no units)', fontsize = 12)
plt.ylabel('Output $y$ (no units)', fontsize = 12)

plt.axhline(0, ls=':',color='k')
plt.axvline(0, ls=':', color='k')

plt.savefig("ReluNoTitle.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
def tanh_deriv(x):
    return 1.0 - np.tanh(x)**2

def logistic_derivative(x):
    return sigmoid(x)*(1-sigmoid(x))

plt.figure(figsize=(5,2.5))
plt.plot(x,tanh_deriv(x), '--', color='silver')
plt.plot(x,logistic_derivative(x), color='k',)
plt.plot(x[:50],[0]*50, color='darkgrey',)
plt.plot(x[50:],[1]*50, color='darkgrey',)
plt.text(-2.2,0.8, r"$\frac{d}{dx}tanh(x)$", fontsize = 10)
plt.annotate(r"$\frac{d}{dx}sigmoid(x)$", xy=(-0.5,0.25), xytext=(-3.5,0.5),
             arrowprops=dict(arrowstyle="->", shrinkA=0), fontsize = 10)
plt.annotate(r"$\frac{d}{dx}\frac{1}{2}tanh(x/2)$", xy=(0.5,0.25), xytext=(1.2,0.5),
             arrowprops=dict(arrowstyle="->", shrinkA=0), fontsize = 10)
plt.text(2,0.85, r"$\frac{d}{dx}ReLU(x)$", fontsize = 10)

plt.grid()

plt.xlim(x_min,x_max)
plt.ylim(-0.2,y_max)

plt.title('Derivatives of Activation Functions',fontsize=14)

plt.xlabel('Input $x$ (no units)', fontsize = 12)
plt.ylabel('Output $y$ (no units)', fontsize = 12)

plt.axhline(0, ls=':',color='k')
plt.axvline(0, ls=':', color='k')

plt.savefig("DerivativesofActivationFunctions.png", bbox_inches='tight', dpi=300)
plt.show()
```
