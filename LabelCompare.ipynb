```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15,5]
plt.plot(y)
plt.plot(yy)
plt.plot(yyy)
plt.show()
```

```python
def onoff(ybits,vth,rrise,rfall):
	vcap = 0
	yout = np.zeros(len(ybits))==1
	for i,y in enumerate(ybits):
		if y:
			vcap += rrise
			if vcap > 1: vcap = 1
		else:
			vcap -= rfall
			if vcap < 0: vcap = 0
		if vcap > vth: yout[i] = True
	return yout
```

```python
voice_labels = np.load('Actual_y.npz')
voice_out = np.load('Out_y.npz')
```

```python
plt.rcParams['figure.figsize'] = [15.8,5]
pos = 6
start = 0 #44100*30*(pos-1)
end   = -1 #44100*30*(pos)
y1o = voice_out['y1o']
y1o = onoff(y1o,0.4,1000/44100,1/44100)
plt.plot(voice_labels['y1'][start:end])
plt.plot(y1o[start:end])
plt.show()
```
