```python
import SchemDraw
import SchemDraw.elements as elm
from SchemDraw import logic

d = SchemDraw.Drawing(unit = 2)
u = d.unit

# Create comparator element
comp = {'base' : elm.OPAMP,
       'paths' : [[[3*u/12,-u/6],[5*u/12,-u/6],[5*u/12,u/6],[7*u/12,u/6]]]
       }
comp_rev = {'base' : elm.OPAMP,
       'paths' : [[[3*u/12,u/6],[5*u/12,u/6],[5*u/12,-u/6],[7*u/12,-u/6]]]
       }
PMOS = {
    'base' : elm.NFET,
    'shapes' : [{'shape':'arrow','start':[0.4,-0.41],'end':[0.6,-0.41],'headwidth':0.3,'headlength':0.2}]
    }
NMOS = {
    'base' : elm.NFET,
    'shapes' : [{'shape':'arrow','start':[0.25,-1.09],'end':[0.05,-1.09],'headwidth':0.3,'headlength':0.2}]
    }
FG_NMOS = {
    'base' : NMOS,
    'paths' : [[[ 0.79166667, -0.41666667],[ 0.79166667, -1.08333333]]]
}
ARC = {
    'shapes' : [{'shape':'arc', 'center':[0,0], 'width':u/4, 'height':u/4, 'theta1' : 0, 'theta2' : 180}],
    'anchors' : {'p1':[u/8, 0],'p2':[-u/8,0]}
}
SW_NOT = {'base' : elm.SWITCH_SPST,
         'shapes' : [{'shape': 'circle', 'center': [0.5,0.4], 'radius': 0.1}]
         }
```

```python
## Code for drawing the envelope detector circuit

# Initialize schematic
d = SchemDraw.Drawing(font = "Times New Roman", unit = 2, inches_per_unit = 0.25, fontsize = 10, lw = 1.2)
u = d.unit

#Draw feedback capacitor into in1 of opamp
d.add(elm.GND_SIG)
d.add(elm.CAP, d = 'right', label = '$C_{FB}$')
dot1 = d.add(elm.DOT)
d.push()
d.add(elm.LINE, d = 'down', l = d.unit/2)
d.add(elm.LINE, d = 'right', l = d.unit/4)
opamp = d.add(comp_rev, d = 'right', anchor = 'in2', flip = True)

# Go back to top of cap, draw R_FB and draw to positive feedback dot
d.pop()
d.add(elm.LINE, l = d.unit/2.5)
d.add(elm.RES, d = 'right', label = '$R_{FB}$')
d.add(elm.LINE, l = d.unit/1.5)
d.add(elm.LINE, d = 'down', l = d.unit*1.6)
dot2 = d.add(elm.DOT)
d.push()

# Add feedback switch and input node
pfet = d.add(PMOS, d = 'down', anchor = 'source', flip = True, reverse = True)
d.add(elm.LINE, d = 'left', xy = pfet.drain, tox = dot1.start)
dot3 = d.add(elm.DOT)
d.add(elm.LINE, d = 'up', toy = opamp.in1)
dot4 = d.add(elm.DOT)
d.push()
d.add(elm.LINE, d = 'right', tox = opamp.in1)
d.pop()
d.add(elm.LINE, d = 'left', l = d.unit/4)
d.add(elm.DOT_OPEN, lftlabel = '$v_{SIG}$')

# Add connection from opamp to switch
d.add(elm.LINE, xy = opamp.out, d = 'right', tox = pfet.gate)
d.add(elm.LINE, d = 'down', toy = pfet.gate)

# Go back to feedback dot, add decay resistor
d.pop()
d.add(elm.LINE, d = 'down', l = d.unit/1.75)
d.add(elm.LINE, d = 'left', l = d.unit/1.5)
d.add(elm.RES, label = '$R_{DECAY}$')
d.add(elm.LINE, tox = dot3.start)
d.add(elm.LINE, d = 'up', toy = dot3.start)
d.add(elm.DOT, xy = [dot2.start[0], opamp.out[1]])

# Add output node and capacitor
d.add(elm.LINE, d = 'right', l = d.unit/2)
d.add(elm.DOT_OPEN, rgtlabel = '$v_{ENV}$')
d.add(elm.CAP, d = 'down', botlabel = '$C_{ENV}$')
d.add(elm.GND_SIG)

d.draw()
d.save('env_detect.png')
```

```python
## Code for drawing the Noise estimator circuit

# Initialize schematic
d = SchemDraw.Drawing(font = "Times New Roman", unit = 2, inches_per_unit = 0.25, fontsize = 10, lw = 1.2)
u = d.unit

#Draw feedback capacitor into in1 of opamp
dot_in = d.add(elm.DOT_OPEN, d = 'right', lftlabel = '$v_{ENV}$')
d.add(elm.LINE, l = u/4)
opamp = d.add(comp, d = 'right', anchor = 'in1')
d.add(elm.LINE, l = u/4, xy = opamp.out)
inv_in = d.add(elm.DOT)

# Add pull-up path
d.push()
d.add(elm.LINE, d = 'up', l = u/2)
d.add(elm.LINE, d = 'right', l = u/8)
pfet = d.add(PMOS, anchor = 'gate', reverse = True)
d.add(elm.RES, d = 'right', xy = pfet.drain, label = '$R_{UP}$')
d.add(elm.VDD)

# Add pull-down path
d.pop()
d.add(elm.LINE, d = 'down', l = u/2)
d.add(elm.LINE, d = 'right', l = u/8)
nfet = d.add(NMOS, anchor = 'gate', reverse = True)
d.add(elm.LINE, d = 'down', xy = nfet.source, l = u/2)
d.add(elm.RES, d = 'left', botlabel = '$R_{DN}$')
d.add(elm.GND_SIG)
d.add(elm.LINE, d = 'down', xy = pfet.source, toy = nfet.drain)

# Add output node and capacitor
d.add(elm.DOT, d = 'right', xy = [nfet.drain[0],inv_in.start[1]],)
d.add(elm.LINE, l = d.unit/2)
dot_fb = d.add(elm.DOT)
d.push()
d.add(elm.LINE, l = d.unit/2)
d.add(elm.DOT_OPEN, rgtlabel = '$v_{NOI}$')
d.add(elm.CAP, d = 'down', botlabel = '$C_{NOI}$')
d.add(elm.GND_SIG)

# Add feedback path
d.pop()
d.add(elm.LINE, d = 'down', l = u)
d.add(elm.LINE, d = 'left', tox = (nfet.drain[0]+u/8))
bump = d.add(ARC, anchor = 'p1', reverse = True, flip = True)
d.add(elm.LINE, d = 'left', xy = bump.p2, tox = dot_in.start)
d.add(elm.LINE, d = 'up', toy = opamp.in2)
d.add(elm.LINE, d = 'right', tox = opamp.in2)

d.draw()
d.save('noise_level.png')
```

```python
d = SchemDraw.Drawing(font = "Times New Roman", unit = 2, inches_per_unit = 0.25, fontsize = 10, lw = 1.2)
u = d.unit

space = 1.25*u
gap = 1.25*u
out_len = 1*u

#d.add(elm.LINE, d = 'left', l = 1.5*u)
m1 = d.add(FG_NMOS,anchor = 'gate', reverse = True)
d.add(elm.LINE, d = 'right', xy = m1.drain, l = space)
d.add(elm.DOT)
m2 = d.add(FG_NMOS, anchor = 'drain', reverse = True)
d.add(elm.LINE, d = 'right', xy = m2.drain, l = space)
d.add(elm.DOT)
d.push()
m3 = d.add(FG_NMOS,anchor = 'drain', reverse = True)
d.add(elm.LINE, d = 'right', xy = m1.source, tox = m2.source)
d.add(elm.DOT)
d.add(elm.LINE, tox = m3.source)
d.add(elm.DOT)
#d.add(elm.LINE, l = u/2)
d.add(elm.GND_SIG, d = 'up')
d.pop()
lo1 = d.add(elm.LINE, d = 'right',l = out_len)
d.labelI_inline(lo1, '$i_{O0}$', d = 'out', start = False, ofst = -u/4)
d.add(elm.DOT_OPEN)


m4 = d.add(FG_NMOS, anchor = 'gate', reverse = True, xy = (m1.gate[0], m1.gate[1]-gap))
d.add(elm.LINE, d = 'right', xy = m4.drain, l = space)
d.add(elm.DOT)
m5 = d.add(FG_NMOS, anchor = 'drain', reverse = True)
d.add(elm.LINE, d = 'right', xy = m5.drain, l = space)
d.add(elm.DOT)
d.push()
m6 = d.add(FG_NMOS,anchor = 'drain', reverse = True)
d.add(elm.LINE, d = 'right', xy = m4.source, tox = m5.source)
d.add(elm.DOT)
d.add(elm.LINE, tox = m6.source)
d.add(elm.DOT)
#d.add(elm.LINE, l = u/2)
d.add(elm.GND_SIG, d = 'up')
d.pop()
lo2 = d.add(elm.LINE, d = 'right', l = out_len)
d.labelI_inline(lo2, '$i_{O1}$', d = 'out', start = False, ofst = -u/4)
d.add(elm.DOT_OPEN)


m7 = d.add(FG_NMOS, anchor = 'gate', reverse = True, xy = (m4.gate[0], m4.gate[1]-gap))
d.add(elm.LINE, d = 'right', xy = m7.drain, l = space)
d.add(elm.DOT)
m8 = d.add(FG_NMOS, anchor = 'drain', reverse = True)
d.add(elm.LINE, d = 'right', xy = m8.drain, l = space)
d.add(elm.DOT)
d.push()
m9 = d.add(FG_NMOS,anchor = 'drain', reverse = True)
d.add(elm.LINE, d = 'right', xy = m7.source, tox = m8.source)
d.add(elm.DOT)
d.add(elm.LINE, tox = m9.source)
d.add(elm.DOT)
#d.add(elm.LINE, l = u/2)
d.add(elm.GND_SIG, d = 'up')
d.pop()
lo3 = d.add(elm.LINE, d = 'right', l = out_len)
d.labelI_inline(lo3, '$i_{O2}$', d = 'out', start = False, ofst = -u/4)
d.add(elm.DOT_OPEN)


d.add(elm.LINE, d = 'up', xy = m7.gate, toy = m4.gate)
d.add(elm.DOT)
d.add(elm.LINE, d = 'up', toy = m1.gate)
d.add(elm.DOT)
d.add(elm.LINE, d = 'up', l = 3*u/4)
d.add(elm.DOT_OPEN, lftlabel = '$v_{I0}$')

d.add(elm.LINE, d = 'up', xy = m8.gate, toy = m5.gate)
d.add(elm.DOT)
d.add(elm.LINE, d = 'up', toy = m2.gate)
d.add(elm.DOT)
d.add(elm.LINE, d = 'up', l = 3*u/4)
d.add(elm.DOT_OPEN, lftlabel = '$v_{I1}$')

d.add(elm.LINE, d = 'up', xy = m9.gate, toy = m6.gate)
d.add(elm.DOT)
d.add(elm.LINE, d = 'up', toy = m3.gate)
d.add(elm.DOT)
d.add(elm.LINE, d = 'up', l = 3*u/4)
d.add(elm.DOT_OPEN, lftlabel = '$v_{I2}$')


d.draw()
d.save('FC_layer.png')
```
