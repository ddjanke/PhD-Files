```python
import pandas as pd
df0 = pd.read_json("random_feature_trials0.json", orient = "index")
df1 = pd.read_json("random_feature_trials1.json", orient = "index")
df = pd.concat([df0, df1], ignore_index = True)
df.head()
```

```python
def dict_to_df(column):
    df = pd.DataFrame.from_dict(column, orient = "index")
    return df


config = dict_to_df(df["config"].to_dict())
accs = dict_to_df(df["accs"].to_dict())
accs.columns = ["acc_train", "acc_val"]
f1 = dict_to_df(df["f1"].to_dict())
f1.columns = ["f1_train", "f1_val"]
```

```python
print(df["max_weight"].max())
```

```python
full_df = config.merge(accs, how = "left", left_index = True, right_index = True) \
                .merge(f1, how = "left", left_index = True, right_index = True) \
                .merge(df["max_weight"], how = "left", left_index = True, right_index = True)
best_df = full_df[full_df["acc_val"] > 0.79]
print(full_df.shape, best_df.shape)
best_df.sort_values(by = ["spread", "f1_val"], axis = 0, ascending = False)
#full_df.head()
```

```python
best_df.to_csv("best_params20200826.csv")
```
